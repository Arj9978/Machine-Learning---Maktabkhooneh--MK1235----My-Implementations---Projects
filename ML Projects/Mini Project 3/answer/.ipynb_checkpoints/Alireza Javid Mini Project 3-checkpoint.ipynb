##### Optimal value of K

```python
error_rate = []
# Might take some time
# for i in range(1,40):
for i in range(1,40,2):      # in as janbe inke adade fard ra peymayesh mikonad behtar ast.
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_KNN,y_train_KNN)
    pred_i = knn.predict(X_test_KNN)
    error_rate.append(np.mean(pred_i != y_test_KNN))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40,2),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=23)
```

```python
knn.fit(X_train_KNN, y_train_KNN)
```

```python
predictions_KNN = knn.predict(X_test_KNN)
```

```python
predictions_KNN
```

#### 3.Decision Tree

```python
X_DT = df.drop(['Survived'], axis=1)
y_DT = df['Survived']
```

```python
X_train_DT, X_test_DT, y_train_DT, y_test_DT = train_test_split(X_DT, y_DT, test_size = 0.25, random_state = 3)
```

```python
sc = StandardScaler()
X_train_DT = sc.fit_transform(X_train_DT)
X_test_DT = sc.transform(X_test_DT)
```

```python
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 3)
classifier.fit(X_train_DT, y_train_DT)
```

```python
predictions_DT = classifier.predict(X_test_DT)
```

```python
predictions_DT
```

```python
cm = confusion_matrix(y_test_DT, predictions_DT)
print(cm)
accuracy_score(y_test_DT, predictions_DT)
```

#### 4.Random Forest

```python
X_RF = df.drop(['Survived'], axis=1)
y_RF = df['Survived']
```

```python
X_train_RF, X_test_RF, y_train_RF, y_test_RF = train_test_split(X_RF, y_RF, test_size = 0.25, random_state = 4)
```

```python
sc = StandardScaler()
X_train_RF = sc.fit_transform(X_train_RF)
X_test_RF = sc.transform(X_test_RF)
```

```python
# Building  Random Forest Classifier
rfc = RandomForestClassifier(criterion = 'entropy', random_state = 4)
rfc.fit(X_train_RF, y_train_RF)

# Evaluating on Training set
rfc_pred_train_RF = rfc.predict(X_train_RF)
print('Training Set Evaluation F1-Score=>',f1_score(y_train_RF,rfc_pred_train_RF))
```

```python
# Evaluating on Test set
rfc_pred_test_RF = rfc.predict(X_test_RF)
print('Testing Set Evaluation F1-Score=>',f1_score(y_test_RF,rfc_pred_test_RF))
```

```python
rfc.feature_importances_
```

```python
feature_importance=pd.DataFrame({'rfc':rfc.feature_importances_},index=df.drop(columns=['Survived']).columns)
feature_importance.sort_values(by='rfc',ascending=True,inplace=True)

index = np.arange(len(feature_importance))
fig, ax = plt.subplots(figsize=(18,8))
rfc_feature=ax.barh(index,feature_importance['rfc'],0.4,color='purple',label='Random Forest')
ax.set(yticks=index+0.4,yticklabels=feature_importance.index)

ax.legend()
plt.show()
```
