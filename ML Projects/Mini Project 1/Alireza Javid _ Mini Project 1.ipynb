## Data Preprocessing

Now I want to delet data's which they have Missing Values.

```python
# Check for missing values
df.isnull()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df = df.dropna(axis=0, how='any', subset=None, inplace=False)
```

```python
# Check for missing values
df.isnull().sum()
```

## Strorytelling - Visualization

```python
sns.displot(df, x="Rating")
```

```python
sns.displot(df, x="Type")
```

```python
sns.displot(df, x="Content Rating", discrete=True, shrink=0.8)
```

```python
sns.displot(df, x="Rating", hue='Type',kind="kde")
```

```python
sns.displot(df, x="Rating", hue='Installs',kind="kde")
```

```python
sns.catplot(x="Rating", y="Installs", data=df)
```

```python
sns.catplot(x="Rating", y="Content Rating", data=df)
```

```python
sns.catplot(x="Rating", y="Installs", kind="box", data=df)
```

```python
sns.catplot(x="Rating", y="Content Rating", kind="box", data=df)
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(x="Rating", y="Installs", data=df)
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(x="Rating", y="Content Rating", data=df)
```

```python
g = sns.FacetGrid(df, col="Installs",  row="Content Rating")
g.map_dataframe(sns.histplot, x="Rating")
g.set_axis_labels("Rating", "Count")
```
