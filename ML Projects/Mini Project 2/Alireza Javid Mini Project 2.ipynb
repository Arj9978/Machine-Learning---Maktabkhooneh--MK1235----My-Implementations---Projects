#### 3.Polynomial regression 

```python
"Creates a polynomial regression model for the given degree"
poly_features = PolynomialFeatures(degree=2)
   
# transform the features to higher degree features.
X_train_quadratic = poly_features.fit_transform(X_train_2)
   
# fit the transformed features to Linear Regression
quadratic = LinearRegression()

quadratic.fit(X_train_quadratic, y_train_2)
     
# predicting on training data-set
y_train_predicted = quadratic.predict(X_train_quadratic)
   
# predicting on test data-set
y_test_predicted = quadratic.predict(poly_features.fit_transform(X_test_2))
```

```python
# Model Train Evaluation
print('R^2(train):',metrics.r2_score(y_train_2, y_train_predicted))
print('MAE(train):',metrics.mean_absolute_error(y_train_2, y_train_predicted))
print('MSE(train):',metrics.mean_squared_error(y_train_2, y_train_predicted))
print('RMSE(train):',np.sqrt(metrics.mean_squared_error(y_train_2, y_train_predicted)))
```

```python
# Model Test Evaluation
print('R^2(test):',metrics.r2_score(y_test_2, y_test_predicted))
print('MAE(test):',metrics.mean_absolute_error(y_test_2, y_test_predicted))
print('MSE(test):',metrics.mean_squared_error(y_test_2, y_test_predicted))
print('RMSE(test):',np.sqrt(metrics.mean_squared_error(y_test_2, y_test_predicted)))
```

```python
"Creates a polynomial regression model for the given degree"
poly_features = PolynomialFeatures(degree=3)
   
# transform the features to higher degree features.
X_train_cubic = poly_features.fit_transform(X_train_2)
# fit the transformed features to Linear Regression
cubic = LinearRegression()

cubic.fit(X_train_cubic, y_train_2)
     
# predicting on training data-set
y_train_predicted = cubic.predict(X_train_cubic)
   
# predicting on test data-set
y_test_predicted = cubic.predict(poly_features.fit_transform(X_test_2))
```

#### 4.KNN Regressor

```python
X_4 = df['temp']
Y_4 = df['count']
```

```python
X_train_4, X_test_4 , y_train_4 , y_test_4 = train_test_split(X_4,Y_4,test_size=0.25,random_state=30)
```

```python
X_train_4 = X_train_4.to_numpy()
X_train_4 = np.reshape(X_train_4, (8164, 1))
y_train_4 = y_train_4.to_numpy()

X_test_4 = X_test_4.to_numpy()
X_test_4 = np.reshape(X_test_4, (2722, 1))
y_test_4 = y_test_4.to_numpy()
```

```python
n_neighbors = 5

for i, weights in enumerate(["uniform", "distance"]):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
    y_ = knn.fit(X_train_4, y_train_4).predict(X_test_4)
    plt.subplot(2, 1, i + 1)
    plt.scatter(X_train_4, y_train_4, color="darkorange", label="data")
    plt.plot(X_test_4, y_, color="navy", label="prediction")
    plt.axis("tight")
    plt.legend()
    plt.title("KNeighborsRegressor (k = %i, weights = '%s')" % (n_neighbors, weights))

plt.tight_layout()
plt.show()
```
