## convert sex and smoker columns to values

```python
df
```

```python
# df.replace({ 'sex': {'Male':0 , 'Female':1} , 'smoker' : {'No': 0 , 'Yes': 1}} ,inplace=True)
# df = df.replace({ 'sex': {'Male':0 , 'Female':1} , 'smoker' : {'No': 0 , 'Yes': 1}} )

# Convert sex and smoker from category → int using map
df["sex"] = df["sex"].map({"Male": 0, "Female": 1}).astype(int)
df["smoker"] = df["smoker"].map({"No": 0, "Yes": 1}).astype(int)

df.head()
```

```python
days = pd.get_dummies(df['day'], dtype=int)  # or np.int8, float, etc.
days.sample(5)
```

```python
days=pd.get_dummies(df['day'], dtype=int)
df = pd.concat([df,days],axis=1)
```

```python
times=pd.get_dummies(df['time'], dtype=int)
df = pd.concat([df,times],axis=1)
```

```python
X = df[["total_bill", "size"]]
y = df["tip"]
```

```python
X
```

```python
y
```

<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/0/05/Scikit_learn_logo_small.svg/390px-Scikit_learn_logo_small.svg.png">

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=26)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
# regressor.fit(X,Y) 
```

```python
predictions = regressor.predict(X_test)
```

```python
X_test
```

```python
predictions
```

```python
y_test
```

```python
y_test.values
```

```python
# y_test is a DataFrame with one column
residuals = np.array(y_test) - np.array(predictions)
residuals
```

```python
print("residuals shape:", residuals.shape)
print("first 10 residuals:", residuals[:10])
print("NaN count:", np.isnan(residuals).sum())
print("finite values:", np.isfinite(residuals).sum())
```

```python
sns.histplot(x=residuals, kde=True, color="blue")  # any color string is fine
plt.xlabel("Residuals (y_test - y_pred)")
plt.title("Residual Distribution")
plt.show()
```

### 1. Mean Absolute Error (MAE) is the mean of the absolute value of the errors. It is calculated as:

<img src="https://cdn-images-1.medium.com/max/800/1*4kvomfLGxysM67hza_-B9Q.png" width=25% />

###  2. Mean Squared Error (MSE) is the mean of the squared errors and is calculated as:

<img src="https://cdn-images-1.medium.com/max/800/1*T37cOEU9OkXNPuqGQcXHSA.png" width=25% />

###  3. Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors:

<img src="https://cdn-images-1.medium.com/max/800/1*SGBsn7WytmYYbuTgDatIpw.gif" width=25% />

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
